```python
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

```python
image = cv2.imread('images/pizza_bluescreen.jpg')
print("This image is:",type(image),"with dimensions:",image.shape)
print("BGR IMAGE")
plt.imshow(image)
```

```python
%matplotlib inline
print("RGB IMAGE")
image_copy = np.copy(image)
image_copy = cv2.cvtColor(image_copy,cv2.COLOR_BGR2RGB)
plt.imshow(image_copy)
```

```python
plt.imshow(image)
```

```python
#Defining Color threshold
lower_blue = np.array([0,0,230])
upper_blue = np.array([50,50,255])
```

```python
#Defining Mask
mask = cv2.inRange(image_copy,lower_blue,upper_blue)
plt.imshow(mask,cmap='gray')
```

```python
masked_image = np.copy(image_copy)
masked_image[mask!=0]=[0,0,0]
plt.imshow(masked_image)
```

```python
space_img = cv2.imread('images\space_background.jpg')
space_img = space_img[0:514,0:816]
space_img = cv2.cvtColor(space_img,cv2.COLOR_BGR2RGB)
plt.imshow(space_img)
```

```python
background_mask = np.copy(space_img)
background_mask[mask==0]=[0,0,0]
plt.imshow(background_mask)
```

```python
final_image = masked_image + background_mask
plt.imshow(final_image)
```
